# Стратегический План компании «Озон Фармацевтика»

## Блок 1. Стратегия

### 1.1. Выбор целевого рынка и продукта

По итогам сравнительного анализа макроэкономических показателей, емкости рынка и уровня конкуренции в фокус-листе стран (Польша, Турция, Вьетнам) наиболее предпочтительным целевым рынком для ОФ является Турция:

- **Объем фармацевтического рынка Турции в 2024 году составил $8,0 млрд. По прогнозам, среднегодовые темпы роста рынка в 2024-2031 гг. составят 4,5%.**
- Около 20% рынка приходится на госпитальные закупки. При этом в стране активно развивается добровольное медицинское страхование (рост на 15% г/г, покрытие до 30% населения к 2030 г.), что создает привлекательный сегмент платежеспособного спроса на инновационные препараты.
- Важный структурный тренд - увеличение доли онкологических заболеваний. По данным Global Cancer Observatory, число новых случаев рака в Турции вырастет со 210 тыс. в 2023 г. до 300 тыс. в 2030 г.
- Наиболее перспективным продуктом ОФ для турецкого рынка является биотехнологический препарат Реналокс (моноклональное антитело) для терапии метастатического почечно-клеточного рака. Патентная защита препарата действует до 2038 года.
- В Турции отсутствует локальное производство биоаналогов в онкологии, а зарегистрированные зарубежные препараты на основе того же механизма действия (авелумаб) имеют цену на 20% выше, чем может обеспечить ОФ при трансфере субстанции и локализации производства готовой формы в Турции.

### 1.2. Конкурентный анализ

Ключевыми конкурентами ОФ на турецком рынке биопрепаратов для терапии мПКР будут Pfizer (Авелума/авелумаб), BMS (Опдиво/ниволумаб) и Eisai (Ленвима/ленватиниб).

- **Pfizer** лидирует по доле на рынке мПКР (около 40%) и узнаваемости бренда (92% онкологов). Компания активно инвестирует в КИ биоаналогов и имеет налаженную систему взаимодействия с лидерами мнения через Advisory Boards.
- Потенциальными конкурентами могут стать **Amgen** и **Dr.Reddy**, которые ведут КИ биоаналогов авелумаба в Турции и могут выйти на рынок в 2025-2026 гг.
- Уникальное преимущество ОФ - инновационный механизм действия препарата (анти-PD-L1 в комбинации с блокадой VEGF), что позволяет достигать большей выживаемости. Ни один из текущих игроков не имеет препарата с такой комбинацией в портфеле.

### 1.3. Способ выхода

Согласно экспертным интервью с представителями Ассоциации международных фармпроизводителей Турции (AIFD), наиболее распространенной и успешной моделью является создание локального производства на базе СП с турецкими компаниями.

В качестве потенциального партнера для ОФ предлагается компания **Eczacıbaşı-Baxter Hastane Ürünleri (EBH)**. Ключевые преимущества:

- Современные мощности и реализованный опыт трансфера технологий от ведущих глобальных фармкомпаний (Baxter, Sanofi).
- Наличие активного фармацевтического ингредиента для авелумаба в портфеле, что облегчает регистрацию.
- Выстроенная сеть дистрибуции в госпитальном сегменте, долгосрочные отношения с администраторами больниц и ключевыми экспертами.

По предварительной оценке, СП с EBH позволит выйти на рынок Турции на **1-1,5 года** быстрее по сравнению со строительством собственного завода, а также на **30%** снизит сумму первоначальных инвестиций, сокращая необходимый объем капитальных затрат с $11 млн до приблизительно $7,7 млн, и обеспечит доступ к ключевым каналам продаж с первого дня запуска.

## Блок 2. Продукт

### 2.1. Ценностное предложение

Для пациентов с метастатическим почечно-клеточным раком препарат Реналокс обеспечивает увеличение медианы общей выживаемости на **6 месяцев** (с 18 до 24 мес.) и снижение риска прогрессирования на **30%** по сравнению со стандартом терапии ниволумабом. При этом профиль безопасности Реналокса характеризуется меньшей частотой тяжелых нежелательных явлений (**10% vs 15%**), особенно со стороны ЖКТ и эндокринной системы. Инновационная лекарственная форма Реналокса (**п/к введение 1 раз в 3 недели** вместо в/в капельниц) обеспечивает большее удобство применения для пациентов и медицинских работников, экономию времени и ресурсов клиник.

#### Ключевые преимущества Реналокса для врачей:

- Увеличение продолжительности и качества жизни пациентов.
- Удобство применения, возможность амбулаторного введения.
- Потенциал комбинированного применения с таргетной терапией и другими ИО-препаратами.
- Управляемый профиль НЯ, низкая частота отмены терапии.

#### Ценность для организаторов здравоохранения и плательщиков:

- Снижение прямых медицинских затрат за счет сокращения госпитализаций, амбулаторных посещений, сопутствующей терапии.
- Улучшение клинических и экономических результатов лечения рака почки в масштабах популяции.
- Повышение престижа клиник за счет внедрения инновационной терапии мирового уровня.

### 2.2. Путь пациента

В результате проведенных фокус-групп и анализа реальной практики в референтных клиниках Стамбула и Анкары выявлено, что текущая модель оказания помощи пациентам с мПКР в Турции характеризуется следующими особенностями и болевыми точками:

- Низкая выявляемость мПКР на ранних стадиях из-за отсутствия скрининга и настороженности терапевтов → поздняя диагностика, ухудшение прогноза.
- Пробелы в знаниях онкологов об эффективности новых опций терапии, низкий уровень междисциплинарного взаимодействия → субоптимальный выбор терапии 1-2 линии.
- Высокая частота госпитализаций для проведения инфузионной терапии, перегрузка коечного фонда и медперсонала → рост расходов, снижение качества жизни пациентов.
- Ограниченное возмещение дорогостоящих таргетных и ИО-препаратов → высокие расходы пациентов out-of-pocket, низкая приверженность терапии.

Для построения оптимального пути пациента с Реналоксом предлагается перепроектировать процесс оказания помощи:

- Повысить онконастороженность и раннюю диагностику мПКР за счет программ обучения терапевтов, создания центров компетенций на базе университетских клиник.
- Стандартизировать алгоритмы выбора терапии 1-2 линии мПКР на основе международных и локальных гайдлайнов с фокусом на комбинированном применении ИО и таргетной терапии.
- Оптимизировать маршрутизацию пациентов с приоритетом амбулаторного введения Реналокса, дистанционного мониторинга состояния, домашней сопроводительной терапии.
- Проработать с государственными органами возможность включения Реналокса в программы возмещения для повышения доступности инновационной терапии.

### 2.3. Готовность платить

По данным Турецкого статистического института (TUIK), расходы на здравоохранение составляют около **5%** от общих расходов домохозяйств. При этом за последние 3 года реальные расходы на лекарства на душу населения снизились на **20%** из-за девальвации лиры и отставания роста доходов от инфляции цен на ЛС.

В условиях универсального государственного страхового покрытия через Фонд социального обеспечения (SGK) пациенты оплачивают из своего кармана в среднем **10-20%** от розничной цены рецептурных препаратов.

Средняя цена одного цикла таргетной терапии мПКР в Турции (сорафениб, сунитиниб, акситиниб) составляет **$500**, в то время как разовая доза Реналокса ex-factory стоит **$1100**.

По результатам опроса 30 пациентов с мПКР, средняя готовность платить out-of-pocket за препарат с профилем эффективности Реналокса составила **$150-200** в месяц при медианереходе расходов домохозяйств на лекарства **$100** в месяц.

Таким образом, для обеспечения финансовой доступности Реналокса критически важно его включение в программы возмещения SGK с учетом механизма ценообразования на инновационные ЛС (референтные цены, переговорные скидки, соглашения о разделении рисков). Необходимый уровень со-оплаты для пациента - не более **$50-80** в месяц.

- **Вторичная упаковка готовой ЛФ** в Турции. На упаковку наносится инструкция на турецком языке, штрих-код Datamatrix для прослеживаемости. Фасовка - по 1 флакону в пачке. Вторичная упаковка валидируется на соответствие требованиям учета лекарств Минздрава Турции.
  
- **Дистрибуция Реналокса** через логистический хаб ОФ-EBH в Стамбуле. Оттуда препарат поставляется в **80 ведущих госпиталей и онкоцентров Турции**, а также в **5 ключевых национальных дистрибьюторов**.

#### Коммерческое продвижение и распределение:

- Для быстрого выведения Реналокса на рынок используются каналы и отношения партнера EBH с госпитальным сегментом. Охват **топ-100 центров** в течение первого года после запуска.
  
- На амбулаторном рынке через собственную сеть ОФ из **20 медпредставителей** обеспечивается покрытие **500 онкологов и 200 центров компетенций**. Для стимулирования выписки препарата применяется система мотивации, медицинских грантов, образовательные мероприятия.
  
- Для продвижения препарата пациентам используется сеть партнерских пациентских организаций из числа членов **KANKO (Ассоциация борьбы с раком почки)**. Реализуются программы повышения осведомленности, скрининга, психологической и информационной поддержки пациентов.

- Критической важностью препаратов для жизни пациентов.
- Покрытием части цены за счет страховых программ.

По оценкам IMS, при повышении цены на **1%** спрос на химиотерапию и таргетные препараты снижается в среднем на **0,2-0,3%**. То есть, **10%** рост цен приводит к падению объемов лишь на **2-3%**.

Для категории ИО-препаратов в силу их инновационности и меньшего охвата страховыми программами эластичность может быть выше. Ожидается, что **10%** рост цен на Реналокс может привести к сокращению объемов на **5-7%**.
